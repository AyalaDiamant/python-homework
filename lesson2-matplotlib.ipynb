תרגיל 5

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x, y1, '-b', label='sin(x)')
plt.plot(x, y2, ':r', label='cos(x)')
plt.scatter(x, y1, c='blue', s=10)  # נקודות לסינוס
plt.title('Line and Scatter on the Same Graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
```

```python
 years = range(2010, 2021)
 consumption_oil = [100, 105, 110, 120, 125, 130, 135, 140, 145, 150, 155]
 consumption_gas = [80, 82, 85, 88, 90, 95, 100, 105, 110, 115, 120]
```

```python
fig, axs = plt.subplots(2,2) 
axs[0, 0].plot(years , consumption_oil , 'tab:blue')
axs[0, 0].set_title('consumption_oil')
axs[0, 1].plot(years , consumption_gas, 'tab:red')
axs[0, 1].set_title('consumption_ga')

# התאמת הגרפים למראה נקי יותר|
for ax in axs.flat:
    ax.label_outer()
```
